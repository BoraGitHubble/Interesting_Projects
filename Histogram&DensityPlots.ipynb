```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

# Histogram
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
plt.hist(data["x"], alpha=0.5)
plt.hist(data["y"], alpha=0.5)
plt.show()
```

```python
# Density Plots
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])
sns.kdeplot(data["x"], fill=True)
sns.kdeplot(data["y"], fill=True)
plt.show()
```

```python
# Both histograms and density plots at once
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])

fig, ax = plt.subplots(figsize=(8, 6))

sns.histplot(data['x'], kde=True, ax=ax, label='x')
sns.kdeplot(data['x'], ax=ax)

sns.histplot(data['y'], kde=True, ax=ax, label='y')
sns.kdeplot(data['y'], ax=ax)

ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.set_title('Histograms and Density Plots')
ax.legend()
plt.show()
```
